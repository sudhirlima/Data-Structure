6. Explain the generator in python and how they are define

A generator is a special type of iterator that yields values one at a time using the yield keyword instead of return.
generators are memory efficient great for large data.
generators pause & resume autometically between yeilds




```python
# Define generator
def countdown(n):
    while n > 0:
        yield n
        n -= 1

# Use generator
for i in countdown(5):
    print(i)

```

5. Create an iterator class in python that generates Fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a

# Example usage:
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

6. Write a generator function in python that yeilds the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp
for num in powers_of_two(5):
    print(num)
```

7. Implement a generator function that reads a file line by line and yeilds each line as a string.

```python
def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.rstrip('\n')  # Remove newline character
for line in read_file_lines('example.txt'):
    print(line)
```

9. Write a python program that uses map() to convert a list of temperature from Celcius to Farenheit.

```python
# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Convert to Fahrenheit using map() and lambda
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))

print("Celsius:", celsius_temps)
print("Fahrenheit:", fahrenheit_temps)
```

10. Create a python program that uses filter() to remove all vowels from a given string.

```python
#str1 = ("a, c, e, g, y, i, o, z, x, f, u")
#type(str1)
#not_vowels = list(filter(lambda x: x not in "aeiouAEIOU", str1))
#print(not_vowels)

def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    return  ''.join(filter(lambda x: x not in vowels, input_str))
# Example usage
text = "a, c, e, g, y, i, o, z, x, f, u!"
no_vowels = remove_vowels(text)

print("Original:", text)
print("Without vowels:", no_vowels)
```

11. Imagine an accounting routine used in a book shop. it works on a list with sublists, write a python program, which returns a list with 2-tuples. Each tuple consists of the order number and the price per item and the quantity. the product should be increased by Rs 10, if the value of the order is smaller than Rs 100.00.

```python
orders = [
    [1001, 15.0, 3, "Python Basics"],
    [1002, 40.0, 3, "Advanced Python"],
    [1003, 8.0, 10, "Fluent Python"]
]

# Transform using map and lambda
adjusted_orders = list(map(
    lambda order: (
        order[0],
        order[1] + 10 if (order[1] * order[2]) < 100 else order[1],
        order[2],
        order[3]
    ),
    orders
))

# Output the result
for entry in adjusted_orders:
    print(entry)
```
